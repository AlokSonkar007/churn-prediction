```python
import numpy as np
import pandas as pd
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.model_selection import train_test_split
import pickle
```

```python
df = pd.read_csv('Churn_Modelling.csv')
df
```

```python
label_encoder_gender = LabelEncoder()
df['Gender'] = label_encoder_gender.fit_transform(df['Gender'])
```

```python
df = df.drop(['RowNumber', 'CustomerId', 'Surname'], axis=1)
```

```python
df.head()
```

```python
one_hot_encoder_geo = OneHotEncoder(handle_unknown='ignore')
geo_encoded = one_hot_encoder_geo.fit_transform(df[['Geography']]).toarray()
geo_encoded_df = pd.DataFrame(geo_encoded, columns=one_hot_encoder_geo.get_feature_names_out(['Geography']))
geo_encoded_df
```

```python
df = pd.concat([df.drop('Geography', axis=1), geo_encoded_df], axis=1)
df
```

```python
X = df.drop('EstimatedSalary', axis=1)
y = df['EstimatedSalary']
```

```python
X
```

```python
y
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
with open('label_encoder_gender_regresssion.pkl', 'wb') as file:
    pickle.dump(label_encoder_gender, file)

with open('one_hot_encoder_geo_regresssion.pkl', 'wb') as file:
    pickle.dump(one_hot_encoder_geo, file)

with open('scaler_regresssion.pkl', 'wb') as file:
    pickle.dump(scaler, file) 
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential  # type: ignore
from tensorflow.keras.layers import Dense  # type: ignore
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard  # type: ignore
import datetime
```

```python
model = Sequential(
    [
        Dense(
            64,
            activation='relu',
            input_shape = (
                X_train.shape[1],  # HL1 connected with input layer
            )
        ),
        Dense(
            32,
            activation='relu'  # HL2
        ),
        Dense(
            1
        )
    ]
)

model.compile(optimizer="Adam", loss="mean_absolute_error", metrics=['mae'])
model.summary()
```

```python
from tensorflow.keras.callbacks import EarlyStopping, TensorBoard  # type: ignore
log_dir = "RegressionLogs/fit/" + datetime.datetime.now().strftime("%Y%m%d-%H%M%S")
tensorboard_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
```

```python
tensorflow_callback = TensorBoard(log_dir=log_dir, histogram_freq=1)
early_stopping_callback = EarlyStopping(monitor='val_loss', patience=20, restore_best_weights=True)
```

```python
history = model.fit(
    X_train,
    y_train,
    validation_data = (
        X_test,
        y_test
    ),
    epochs = 100,
    callbacks = [
        tensorflow_callback,
        early_stopping_callback
    ]
)
history
```

```python
model.save('regression_model.h5')
```

```python
test_loss, test_mae = model.evaluate(X_test, y_test)
print(f'Test MAE: ', {test_mae})
```
